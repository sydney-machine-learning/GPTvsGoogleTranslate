```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
import pandas as pd

# Define the paths to your CSV files
csv_paths = ['GeminiSanskrit.csv', 'Gpt4Sanskrit.csv', 'GptSanskrit.csv', 'GoogleTranslateSanskrit.csv', 'OfficialTranslation(Purohit).csv']
csv_names = ['GeminiSanskrit', 'Gpt4Sanskrit', 'GptSanskrit', 'GoogleTranslator', 'OfficialTranslation(Purohit)']

# Read all CSV files into a dictionary of DataFrames
dfs = {csv_name: pd.read_csv(csv_path) for csv_name, csv_path in zip(csv_names, csv_paths)}

# Find common sloka numbers among all DataFrames
common_sloka_numbers = set(dfs[csv_names[0]]['slokas_no'].tolist())
for df in dfs.values():
    common_sloka_numbers.intersection_update(set(df['slokas_no'].tolist()))

# Create a new DataFrame with slokas from all CSV files
combined_df = pd.DataFrame()

# Iterate through common sloka numbers
for sloka_no in common_sloka_numbers:
    sloka_row = {'slokas_no': sloka_no}
    for csv_name, df in dfs.items():
        sloka = df.loc[df['slokas_no'] == sloka_no, 'slokas'].values
        if len(sloka) > 0:
            sloka_row[csv_name] = sloka[0]
        else:
            sloka_row[csv_name] = ''
    combined_df = pd.concat([combined_df, pd.DataFrame([sloka_row])], ignore_index=True)

# Print the combined DataFrame
print(combined_df)
```

```python
sentences = []
sentence5 = combined_df['OfficialTranslation(Purohit)'].tolist()
sentence1 = combined_df['GeminiSanskrit'].tolist()
sentence2 = combined_df['Gpt4Sanskrit'].tolist()
sentence3 = combined_df['GptSanskrit'].tolist()
sentence4 = combined_df['GoogleTranslator'].tolist()

sentences.append(sentence5)
sentences.append(sentence1)
sentences.append(sentence2)
sentences.append(sentence3)
sentences.append(sentence4)


sentence_new = []
for i in range(0, len(sentences)):
    for j in range(0, len(sentences[i])):
        sentence_new.append(sentences[i][j])

#sentence_embeddings = model.encode(sentence_new)
print(sentence_new)
```

```python
sentence_embeddings = model.encode(sentence_new)
print(sentence_embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


i = 0
j = 43
k = 86
l = 129
m = 172
sloka_no = 1
data=[]
for i in range(0, 43):

  gemini_purohit = cosine_similarity([sentence_embeddings[j]], [sentence_embeddings[i]])[0][0]
  gpt4_purohit = cosine_similarity([sentence_embeddings[k]], [sentence_embeddings[i]])[0][0]
  gpt_purohit = cosine_similarity([sentence_embeddings[l]], [sentence_embeddings[i]])[0][0]
  purohit_google = cosine_similarity([sentence_embeddings[m]], [sentence_embeddings[i]])[0][0]

  dict1 = {'Sloka': sloka_no,
           'OfficialTranslation':sentence_new[i],
           "Gemini" : sentence_new[j],
           "GPT 4" :sentence_new[k],
           "Chat-GPT":sentence_new[l],
           "Google Translate":sentence_new[m],
         'Gemini - Purohit': gemini_purohit,
         'Gpt4Sanskrit - Purohit': gpt4_purohit,
         'GptSanskrit - Purohit': gpt_purohit,
         'Google- Purohit ': purohit_google}
  data.append(dict1)
  i+=1
  j+=1
  k+=1
  l+=1
  m+=1
  sloka_no+=1

  df_new = pd.DataFrame(data)
  df_new.to_csv('SentimentSanskrit.csv', index=False)
  #df = pd.concat([df, df_new], ignore_index = True)
```
