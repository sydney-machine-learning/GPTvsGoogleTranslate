```python
import csv

data = {
  "Sentiment": [
    "Optimistic", "Thankful", "Empathetic", "Pessimistic", "Anxious",
    "Sad", "Annoyed", "Denial", "Surprise", "Joking"
  ],
  "OfficialTranslation":[8, 300, 430, 480 ,416, 200, 335,400 ,115,88],
  "Google_Translate":   [0,416, 520, 534,514, 360, 401,389,34, 0],
    "Chat_GPT_3.5":     [35, 325, 450, 620, 356,322, 287, 420, 63, 28],
    "GEMINI":           [20, 490, 416, 390, 424, 368,532, 530, 76, 0],
    "Chat_GPT_4.0":     [55, 580, 480, 456, 492, 288, 387, 321, 140, 38]
   
}

with open('TeluguSentiments.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(data.keys())  # Write header row
  # Write data rows
  for i in range(len(data['Sentiment'])):
      row = [data['Sentiment'][i]]
      for key, value in data.items():
          if key != 'Sentiment':
              row.append(value[i])
      writer.writerow(row)

print("CSV file created successfully!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("SanskritSentiments.csv")

# Extract sentiment categories and model names
sentiment_categories = df['Sentiment'].tolist()
model_names = df.columns[1:].tolist()

# Extract data values (counts) for each model
data_values = [df[model].tolist() for model in model_names]

# Set bar width and create the figure and axis
bar_width = 0.2
fig, ax = plt.subplots(figsize=(18, 10))

# Number of models
num_models = len(model_names)

# Create x positions for each group based on number of models
x = np.arange(len(sentiment_categories))

# Define a color scheme (optional)
colors = ['blue', 'orange', 'green', 'red', "brown"]

# Create grouped bars for each model
for i, model in enumerate(model_names):
  ax.bar(x + i * bar_width, data_values[i], width=bar_width-0.02, label=model, color=colors[i])

# Customize the chart
ax.set_xlabel('Sentiments', fontsize=18)
ax.set_ylabel('Count', fontsize=18)
ax.set_xticks(x + (len(model_names) / 2) * bar_width)
ax.set_xticklabels(sentiment_categories, rotation=45,fontsize=16)
ax.legend(title='Models', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.legend(title='Models',loc='upper right',prop={"size":12})
plt.figure(facecolor='lightblue')
plt.tight_layout()



# Show the plot
plt.show()
```
