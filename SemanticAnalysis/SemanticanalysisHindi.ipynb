```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
import pandas as pd


# Define the paths to your CSV files
csv_paths = ['GeminiHindi.csv', 'Gpt4Hindi.csv', 'GptHindi.csv', 'GoogleTranslateHindi.csv', 'TamasSaini.csv']
csv_names = ['GeminiHindi', 'Gpt4Hindi', 'GptHindi', 'GoogleTranslateHindi', 'TamasSaini']

# Read all CSV files into a dictionary of DataFrames
dfs = {csv_name: pd.read_csv(csv_path) for csv_name, csv_path in zip(csv_names, csv_paths)}
# Find common sloka numbers among all DataFrames
common_sloka_numbers = set(dfs[csv_names[0]]['sentence_no'].tolist())
for df in dfs.values():
    common_sloka_numbers.intersection_update(set(df['sentence_no'].tolist()))

# Create a new DataFrame with slokas from all CSV files
combined_df = pd.DataFrame()

# Iterate through common sloka numbers
for sloka_no in common_sloka_numbers:
    sloka_row = {'sentence_no': sloka_no}
    for csv_name, df in dfs.items():
        sloka = df.loc[df['sentence_no'] == sloka_no, 'text'].values
        if len(sloka) > 0:
            sloka_row[csv_name] = sloka[0]
        else:
            sloka_row[csv_name] = ''
    combined_df = pd.concat([combined_df, pd.DataFrame([sloka_row])], ignore_index=True)

# Print the combined DataFrame
print(combined_df)
```

```python
sentences = []
sentence5 = combined_df['TamasSaini'].tolist()
sentence1 = combined_df['GeminiHindi'].tolist()
sentence2 = combined_df['Gpt4Hindi'].tolist()
sentence3 = combined_df['GptHindi'].tolist()
sentence4 = combined_df['GoogleTranslateHindi'].tolist()

sentences.append(sentence5)
sentences.append(sentence1)
sentences.append(sentence2)
sentences.append(sentence3)
sentences.append(sentence4)


sentence_new = []
for i in range(0, len(sentences)):
    for j in range(0, len(sentences[i])):
        sentence_new.append(sentences[i][j])

#sentence_embeddings = model.encode(sentence_new)
print(sentence_new)
```

```python
sentence_embeddings = model.encode(sentence_new)
print(sentence_embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


i = 0
j = 51
k = 103
l = 155
m = 207
sentence_no = 1
data = []


for i in range(0,51):

  gemini_saini = cosine_similarity([sentence_embeddings[j]], [sentence_embeddings[i]])[0][0]
  gpt4_saini = cosine_similarity([sentence_embeddings[k]], [sentence_embeddings[i]])[0][0]
  gpt_saini = cosine_similarity([sentence_embeddings[l]], [sentence_embeddings[i]])[0][0]
  saini_google = cosine_similarity([sentence_embeddings[m]], [sentence_embeddings[i]])[0][0]

  dict1 = {'Sentence_no': sentence_no ,
           "OfficialTranslation" :sentence_new[i],
           "Gemini" : sentence_new[j],
           "GPT 4" :sentence_new[k],
           "Chat-GPT":sentence_new[l],
           "Google Translate":sentence_new[m],
         'Gemini - Saini': gemini_saini,
         'Gpt4Sanskrit - Saini': gpt4_saini,
         'GptSanskrit - Saini': gpt_saini,
         'Google- Saini ': saini_google}
  data.append(dict1)
  i+=1
  j+=1
  k+=1
  l+=1
  m+=1
  sentence_no+=1

df_new = pd.DataFrame(data)
print(df_new)
df_new.to_csv('SentimentHindi.csv', index=False)
```
